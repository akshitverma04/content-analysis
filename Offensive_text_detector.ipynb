```python
from transformers import pipeline
```

```python
hate_detector = pipeline(
    "text-classification",
    model="Falconsai/offensive_speech_detection"
)
```

```python
nsfw_detector = pipeline(
    "text-classification",
    model="eliasalbouzidi/distilbert-nsfw-text-classifier"
)
```

```python
samples = [
    "I hate you and your kind!",
    "Check out these explicit details on adult content.",
    "What a wonderful day to help others."
]
```

```python
for text in samples:
    hate_result = hate_detector(text)[0]
    nsfw_result = nsfw_detector(text)[0]
    print(f"Text: {text!r}")
    print(f"  Hate/Offensive ➔ {hate_result['label']} (score: {hate_result['score']:.3f})")
    print(f"  NSFW           ➔ {nsfw_result['label']} (score: {nsfw_result['score']:.3f})")
    print("-" * 60)
```
